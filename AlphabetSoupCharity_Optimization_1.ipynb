## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(['EIN', 'NAME', 'SPECIAL_CONSIDERATIONS', 'ORGANIZATION'], axis=1, inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
unique_values
```

```python
# Binning ask_amt column

bin_edges = [5000, 100000, 500000, 1000000, 5000000, 10000000, 500000000, 1000000000, 8597806340] # creating bins based on min ask_amt and max ask_amt
application_df['ASK_AMT_BINNED'] = pd.cut(application_df['ASK_AMT'], bins=bin_edges, right=False)
application_df.head()
```

```python
# Dropping original ASK_AMT column
application_df.drop(['ASK_AMT'], axis=1, inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
unique_values
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
value_counts_application_type = application_df['APPLICATION_TYPE'].value_counts()
print(value_counts_application_type)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
application_types_to_replace = list(value_counts_application_type[value_counts_application_type < cutoff_value].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()

```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
value_counts_classification = application_df['CLASSIFICATION'].value_counts()
print(value_counts_classification)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
value_counts_classification1 = value_counts_classification[value_counts_classification > 1]
print(value_counts_classification1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value_classifications = 1000
classifications_to_replace = list(value_counts_classification[value_counts_classification < cutoff_value_classifications].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Review the info
application_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Review the info
application_df_dummies = pd.get_dummies(application_df, columns=[
    'APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
    'INCOME_AMT', 'ASK_AMT_BINNED'
])

# Display the transformed data
application_df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_dummies.IS_SUCCESSFUL.values
X = application_df_dummies.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

number_of_features = X_train_scaled.shape[1]
nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=number_of_features))

# Additional hidden layers
nn_model.add(tf.keras.layers.Dense(units=50, activation='relu'))
nn_model.add(tf.keras.layers.Dense(units=20, activation='relu'))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
import os
from keras.models import Sequential

#  renaming the model for better understanding
model = nn_model

# Defining the path for saving the file
output_dir = '/content/Output'

# Creating the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Saving the model to an HDF5 file
model.save(os.path.join(output_dir, 'AlphabetSoupCharity1.h5'))
```
